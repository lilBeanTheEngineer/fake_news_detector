```python
def remove_stop_words(articles):
    """
    Removes all the stop words from the text 
    in each articles
    """
    text_without_stop_words = []
    
    for article in articles:
        text_without_stop_words.append([word for word in articles if word not in stop_words])
    
    return text_without_stop_words
```
